```python
from transformers import BertTokenizer, BertModel
from transformers import DistilBertTokenizer, DistilBertModel
from transformers import RobertaTokenizer, RobertaModel
import torch
import scipy

tokenizer = [BertTokenizer.from_pretrained("bert-base-uncased"), RobertaTokenizer.from_pretrained('roberta-base'), DistilBertTokenizer.from_pretrained('distilbert-base-uncased')]

model_bert = [BertModel.from_pretrained("bert-base-uncased"), RobertaModel.from_pretrained('roberta-base'), DistilBertModel.from_pretrained('distilbert-base-uncased')]

for i in range(len(tokenizer)):
  model_name.append(model_bert[i].config._name_or_path)
  for j in range(len(question_1)):

    # Tokenize and encode the sentences
    inputs1 = tokenizer[i](correct_answer1_1[j], return_tensors='pt')
    inputs2 = tokenizer[i](correct_answer2_1[j], return_tensors='pt')
    inputs3 = tokenizer[i](wrong_answer1_1[j], return_tensors='pt')
    inputs4 = tokenizer[i](wrong_answer2_1[j], return_tensors='pt')

    # Generate BERT embeddings
    with torch.no_grad():
        outputs1 = model_bert[i](**inputs1)
        outputs2 = model_bert[i](**inputs2)
        outputs3 = model_bert[i](**inputs3)
        outputs4 = model_bert[i](**inputs4)

    # Get the embeddings from BERT's last hidden states
    embeddings1 = outputs1.last_hidden_state.mean(dim=1).squeeze().numpy()
    embeddings2 = outputs2.last_hidden_state.mean(dim=1).squeeze().numpy()
    embeddings3 = outputs3.last_hidden_state.mean(dim=1).squeeze().numpy()
    embeddings4 = outputs4.last_hidden_state.mean(dim=1).squeeze().numpy()

    # Calculate cosine similarity between embeddings
    similarity_score1 = 1 - scipy.spatial.distance.cosine(embeddings1, embeddings2)
    similarity_score2 = 1 - scipy.spatial.distance.cosine(embeddings1, embeddings3)
    similarity_score3 = 1 - scipy.spatial.distance.cosine(embeddings1, embeddings4)
    similarity_score4 = 1 - scipy.spatial.distance.cosine(embeddings2, embeddings3)
    similarity_score5 = 1 - scipy.spatial.distance.cosine(embeddings2, embeddings4)
    #print("Similarity Score:", similarity_score)

    sentancetransformer_cosinesimilarity_1.append([model_bert[i].config._name_or_path, j ,similarity_score1, similarity_score2, similarity_score3, similarity_score4, similarity_score5])


  # print("\nSimilarity Scores using BERT model and tokenizer:")
  # print("Sentences 1 and 2:", similarity_score1)
  # print("Sentences 1 and 3:", similarity_score2)
  # print("Sentences 2 and 3:", similarity_score3)
  # print("Sentences 3 and 4:", similarity_score4)
```
